```python
#Step 1: Create Empty Student List

students = []
```

```python
#Step 2: Create Add Student Function
#Purpose: Add new student details

def add_student():

    roll = int(input("Enter Roll No: "))
    name = input("Enter Name: ")
    marks = float(input("Enter Marks: "))

    student = {
        "roll": roll,
        "name": name,
        "marks": marks
    }

    students.append(student)

    print("Student Added Successfully")
```

```python
#Step 3: Create View Students Function
#Purpose: Display all records

def view_students():

    if len(students) == 0:
        print("No Records Found")

    else:

        for student in students:

            print(
                student["roll"],
                student["name"],
                student["marks"]
            )
```

```python
#Step 4: Create Search Function
#Purpose: Search by Roll Number

def search_student():

    roll = int(input("Enter Roll No: "))

    found = False

    for student in students:

        if student["roll"] == roll:

            print(student)

            found = True

    if found == False:
        print("Student Not Found")
```

```python
#Step 5: Create Update Function
#Purpose: Modify student details

def update_student():

    roll = int(input("Enter Roll No: "))

    for student in students:

        if student["roll"] == roll:

            student["name"] = input("Enter New Name: ")

            student["marks"] = float(
                input("Enter New Marks: ")
            )

            print("Updated Successfully")

            return

    print("Student Not Found")
```

```python
#Step 6: Create Delete Function
#Purpose: Remove a student

def delete_student():

    roll = int(input("Enter Roll No: "))

    for student in students:

        if student["roll"] == roll:

            students.remove(student)

            print("Deleted Successfully")

            return

    print("Student Not Found")
```

```python
#Step 7: Create Main Menu

while True:

    print("\n===== MENU =====")

    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Exit")

    choice = int(input("Enter Choice: "))
    
#Step 8: Connect Menu with Functions

    if choice == 1:
        add_student()

    elif choice == 2:
        view_students()

    elif choice == 3:
        search_student()

    elif choice == 4:
        update_student()

    elif choice == 5:
        delete_student()

    elif choice == 6:

        print("Thank You")

        break

    else:

        print("Invalid Choice")
```

```python
#Step 9: Save Student Data
#Replace this part in add_student():

def add_student():

    roll = int(input("Enter Roll No: "))
    name = input("Enter Name: ")
    marks = float(input("Enter Marks: "))

    student = {
        "roll": roll,
        "name": name,
        "marks": marks
    }

    students.append(student)

    file = open("students.txt", "a")

    file.write(
        str(roll) + "," +
        name + "," +
        str(marks) + "\n"
    )

    file.close()

    print("Student Added Successfully")
```
